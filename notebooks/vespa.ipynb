```python
from vespa.application import Vespa
import pandas as pd
from requests import HTTPError

app = Vespa("http://localhost", 8080)

print(app.get_application_status())

try:
    response = app.get_data(schema="report", data_id="abc")
    print(response.get_json())
except HTTPError as err:
    print(err.response.status_code)
    print("Document not found.")

data = {
    "organizations": ["allrad"],
    "pacs_aet": "foobar",
    "pacs_name": "foobar",
    "patient_id": "foobar",
    "year_of_birth": 1950,
    "gender": "M",
    "study_uid": "foobar",
    "accession_number": "foobar",
    "study_description": "foobar",
    "study_datetime": 1234,
    "series_uid": "foobar",
    "modalities": ["CT", "PET"],
    "instance_uid": "foobar",
    "references": ["foobar"],
    "body": "foo bar mar",
}

result = app.feed_data_point(schema="report", data_id="abc", fields=data)
print(result.get_json())
```
